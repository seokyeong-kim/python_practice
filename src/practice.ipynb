```python
# 오늘의 할일 관리 프로그램

todo_list = []

while True:
    choice = input("메뉴선택(1:추가,2:삭제,0:종료)? ")

    if choice == "1":
        task = input("* 할일 내용 입력>> ")
        todo_list.append(task)
        print("******************************")
        for i, t in enumerate(todo_list, start=1):
            print(f"{i}.{t}")
        print("******************************")

    elif choice == "2":
        task = input("* 완료한 내용입력>> ")
        if task in todo_list:
            todo_list.remove(task)
        else:
            print(f"'{task}'는 목록에 없습니다.")
        print("******************************")
        for i, t in enumerate(todo_list, start=1):
            print(f"{i}.{t}")
        print("******************************")

    elif choice == "0":
        print("프로그램을종료합니다.")
        break

    else:
        print("메뉴선택오류입니다.다시선택하세요.")
```

```python
todos = []

while True:
    selected = int(input('메뉴 선택(1:추가, 2:삭제, 0:종료)? '))
    if selected == 0:
        print('종료합니다.')
        break

    elif selected == 1:
        print('할일 내용 입력 >>')
        todo = input('입력하세요.')
        todos.append(todo)

    elif selected == 2:
        print('완료할 내용 입력 >>')
        todo = input('입력하세요.')
        todos.remove(todo)

    else:
        print('입력 오류. 다시 입력하세요.')
        continue

    idx = 1
    for value in todos:
       print(f'{idx}. {value}')
       idx += 1
```

```python
students = [
{'name': 'Alice', 'grades': [90, 85, 92]},
{'name': 'Bob', 'grades': [78, 80, 88]},
{'name': 'Charlie', 'grades': [95, 92, 89]}
]

for student in students:
    name = student['name']
    grades = student['grades']
    avg = sum(grades) / len(grades)   # 평균 계산
    print(f"{name}의 평균 성적: {avg:.2f}")



# 모든 학생의 성적을 하나의 리스트로 합치기
all_grades = []
for student in students:
    all_grades.extend(student['grades'])  # 각 학생의 성적을 전체 리스트에 추가

# 전체 평균 계산
overall_avg = sum(all_grades) / len(all_grades)

print(f"전체 학생의 평균 성적: {overall_avg:.2f}")



# 성적이 90 이상인 학생 찾기
high_achievers = []

for student in students:
    # 학생의 성적 중 하나라도 90 이상이면 리스트에 추가
    if any(grade >= 90 for grade in student['grades']):
        high_achievers.append(student['name'])

print("성적이 90 이상인 학생들:", high_achievers)
```

```python
scores = [
    {'Alice': [90, 85, 92]},
    {'Bob': [78, 80, 88]},
    {'Charlie': [95, 92, 89]}
]

averages = []
for kvp in scores:
    for name, grades in kvp.items():
        average = sum(grades) / len(grades)
        averages.append(average)
        print(f"{name}'s average grade: {average:.2f}")
        
print(f"\nAverage of all students: {sum(averages)/len(averages):.2f}")
```

```python
#여러 개의 입력값을 받는 함수 만들기
def add_many(*args):
    total = 0
    for i in args:
        total += i
    return total

result = add_many(1, 2, 3, 4, 5)
print(result)
```

```python
def add_mul(choice, *args):
    if choice == 'add':
        total = 0
        for i in args:
            total += i
        return total
    elif choice == 'mul':
        total = 1
        for i in args:
            total *= i
        return total
    
result2 = add_mul('mul', 1,2,3,4)
print(result2)
```

```python
import random

sel = ['가위', '바위', '보']

def checkWin(user, com):
    if user == com:
        print(f'비겼습니다(사용자: {user}, 컴퓨터: {com})')
        return False
    
    elif (user == '가위' and com == '보') or \
         (user == '바위' and com == '가위') or \
         (user == '보' and com == '바위'): 
         print(f'이겼습니다(사용자: {user}, 컴퓨터: {com})')
         return True 

    else:
        print(f'졌습니다(사용자: {user}, 컴퓨터: {com})')
        return False

def continueConfirm():
    answer = input('게임을 계속하시겠습니까? (y/n): ')
    return answer.lower() == 'y'

while True:
    user = input('가위, 바위, 보: ')
    com = sel[random.randint(0, 2)]

    if checkWin(user, com): 
        if continueConfirm():
            continue
        else:
            break      
```

```python

import random

sel = ['가위', '바위', '보']
dictionary = {}

for idx, value in enumerate(sel):
    dictionary[value] = idx

def checkWin(user, com):
    user_index = dictionary[user]
    com_index = dictionary[com]
    com_index += 1
    com_index = com_index % 3
    if user_index == com_index:
        print('You win')
        return True  
    elif user == com:
        print('비겼습니다')
        return False
    else:
        print('졌습니다')
        return False
    
def continueConfirm():
    print('게임을 계속하시겠습니까? (y/n): ')
    answer = input()
    if answer[0] == 'Y' or answer[0] == 'y':
        return True
    else:
        return False

while True:
    user = input('가위, 바위, 보: ')
    com = sel[random.randint(0, 2)]
    print(f'내가 낸 수 {user}')
    print(f'컴퓨터가 낸 수 {com}')
    if checkWin(user, com): 
        if continueConfirm():
            continue
        else:
            break   
```

```python
#클래스로 계산기 만들기

result1 = 0
result2 = 0

def add1(num): #계산기1
    global result1
    result1 +- num
    return result1
def add2(num):
    global result2
    result2 +- num
    return result2

print(add1(3))
print(add1(4))
print(add2(3))
print(add2(7))


class Calculator:
    def __init__(self):
        self.result = 0
    
    def add(self, num):
        self.result +- num
        return self.result
    
cal1 = Calculator()
cal2 = Calculator()

print(cal1.add(3))
print(cal1.add(4))
print(cal2.add(3))
print(cal2.add(7))
```

```python
class FourCal: 
    def __init__(self, first, second):
        self.first = first
        self.second = second
    def add(self):
        result = self.first + self.second
        return result
    def mul(self):
        result = self.first * self.second
        return result
    def sub(self):
        result = self.first - self.second
        return result
    def div(self):
        result = self.first / self.second
        return result
```

```python
class MoreFourCal(FourCal):
    def pow(self):
        result = self.first ** self.second
        return result
   # def div(self):
        if self.second == 0:
            return 0 
        
        result = self.first / self.second
        return result
```

```python
a = MoreFourCal(4,2)
print(a.add())
print(a.pow())
```

```python
#upper/lower/swapcase -> 크롤링할때 꼭 필요한 함수들

print('Hello World'.upper())
print('Hello World'.lower())
print('Hello World'.swapcase())
```

```python
#trim (공백 잘라주는 함수) -> 파이썬에서는 strip 함수로 호출한다
print(len(' Hello, World!  '))
print(len(' Hello, World!  '.lstrip()))
print(len(' Hello, World!  '.rstrip()))
print(len(' Hello, World!  '.strip()))
```

```python
#split - 자른다. 토큰화한다. Tokenizer
print('Hello, World!'.split())
print('Hello, World! '.split())
print('Hello,  World!'.split())
print('  Hello, World!'.split())

#스페이스는 문자화하지 않는다
```

```python
#JSON

import json

data = {
    "name": "홍길동",
    "age": 30,
    "city": "Seoul"
}  #사실상 파이썬 딕셔너리 데이터

#딕셔너리를 JSON 문자열로 변환
print(json.dumps(data, ensure_ascii=True))
json_string = json.dumps(data, ensure_ascii=False)  #특정 object를 json 형식으로 변환한다는 뜻 #아스키코드 True는 영문, False는 한국어 
print(f'JSON 문자열 : {json_string}')
print(type(json_string))

#JSON 문자열을 다시 파이썬 딕셔너리 데이터 타입으로 변환
parsed_data = json.loads(json_string)
print(f'\n파이썬 데이터 타입 : {type(parsed_data)}')
print(f'\n파이썬 데이터 : {parsed_data}')
```

```python
#파일 '쓰기 기능'으로 열기 

with open('../data/example2.json', 'w', encoding = 'utf-8') as f: #w는 쓰기 전용으로 열겠다는 뜻. r은 read 읽기 전용 #as는 alias의 약자다
    json.dump(data, f, indent=4, ensure_ascii=False)
```

```python
#파일 읽기 기능으로 열기
with open('../data/example2.json', 'r', encoding = 'utf-8') as f:
    content = json.load(f)
    print(content)
    print(type(content))
```

```python
import csv

#name,age,city
#홍길동,30,Seoul
#김철수,25,LA
#이영희,28,Tokyo

csv_data = [
    ['name','age','city'],
    ['홍길동',30,'Seoul'],
    ['김철수',25,'LA'],
    ['이영희',28,'Tokyo']
]

#파일 쓰기
with open('../data/example1.csv', 'w', newline='', encoding = 'utf-8-sig') as f:
    writer = csv.writer(f)
    writer.writerows(csv_data) #csv_data 리스트의 모든 행을 한번에 디스크에 씀 #rows가 row보다 더 빠른 계산함
    #한줄씩 쓰려면 writer.writerow(row)를 사용한다

#데이터를 어디에서 다운로드받든 데이터 전처리를 해야 한다. 만약 미국 데이터를 받았다, 하면 단위를 피트로 쓰니까 미터법으로 변환해야 한다. 
#온도도 화씨를 섭씨로 변환해야 한다. 원본 데이터를 어딘가에 저장을 해야 한다. 그리고 전처리 마치면 원본과 대조하여 잘 됐는지 확인할 것. 

```
