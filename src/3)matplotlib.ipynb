```python
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun gothic')
plt.rc('axes', unicode_minus=False)
```

```python
age = [7,8,9,10,11]
height = [122.1, 127.9, 133.4, 138.8, 142.7]

plt.plot(age, height, color ='red', marker = 'o')
plt.show()
```

```python
blood_type = ['A','B','O','AB']
count = [45, 35, 15, 5]

plt.bar(blood_type, count, color = ['red','blue','green','purple'])
plt.show()
```

```python
plt.barh(blood_type, count, color = ['red','blue','green','purple'])  #차트를 수평으로 만들어준다 barh horizontal
plt.show()
```

```python

plt.pie(count, colors = ['red','blue','green','purple'])  #얘는 color가 아니라 colors라 해야 한다. 
plt.show()
```

```python
plt.pie(count, labels=blood_type, autopct='%.1f%%')  #%.1f%% 은 비율 표시 
plt.show()
```

```python
import seaborn as sns

tips = sns.load_dataset('tips')
tips.head(3)
```

```python
#산점도 그래프 그리기

x = tips['total_bill']
y = tips['tip']
plt.scatter(x,y)
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.show()
```

```python
#히스토그램 그리기

plt.hist(tips['total_bill'])
plt.show()
```

```python
plt.hist(tips['total_bill'], bins = 20)
plt.show()
```

```python
plt.boxplot(tips['total_bill'])
plt.show()

#중간에 노란선이 평균. 위쪽 horizontal line은 75%
```

```python
#샘플데이터

import pandas as pd

df1 = pd.DataFrame({'월':[1,2,3,4,5], '몸무게':[80,78,75,73,70]})
df2 = pd.DataFrame({'월':[1,2,3,4,5], '몸무게':[60,62,59,55,54]})
```

```python
plt.plot(df1['월'], df1['몸무게'], label = '철수')
plt.plot(df2['월'], df2['몸무게'], label = '영희')
plt.legend()
plt.show()
```
