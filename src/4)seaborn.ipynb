```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')  #팁을 얼마나 내는지에 대한 데이터다 
tips.head(10)
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun gothic')
plt.rc('axes', unicode_minus=False)
```

```python
sns.barplot(data=tips, x='day',y='tip',estimator='mean',errorbar=None)
plt.show()
```

```python
sns.barplot(data=tips, x='day',y='tip',estimator='sum',errorbar=None, hue='time', palette='Set1')
plt.show()
```

```python
#산점도
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='smoker', size='size')   #점의 크기가 크면 사이즈가 큰 것이다
plt.show()
```

```python
tips.info()
```

```python
#선형회귀선 linear regression (plot- 그래프를 그리다). 선형의 경향성 찾고, 회귀는 수렴한다는 의미

sns.regplot(data=tips, x='total_bill', y='tip')
plt.show()

값들의 토탈 합이 가장 적은 선을 그리는 것.. 
```

```python
#Count Plot 빈도 막대그래프

tips.head(3) #위쪽 3개만 보여주기
```

```python
tips['size'].value_counts()
```

```python
sns.countplot(data=tips, x='size')  #사이즈를 보게 된다 
plt.show()
```

```python
#Box Plot 상자 수염 그래프. 수치형 데이터의 분포와 이상치를 한 눈에 보여준다. 
sns.boxplot(data=tips, x='size', y='tip')
```

```python
#Box plot with hue and plaette
sns.boxplot(data=tips, x='size', y='tip', hue='time',palette='Set3')
plt.show()
```

```python
#바이올린플롯

sns.violinplot(data=tips, x='size', y='tip')
plt.show()
```

```python
sns.violinplot(data=tips, x='size', y='tip', hue='size')
plt.show()
```

```python
#히스토그램

sns.histplot(data=tips, x='tip', bins=20)
plt.show()
```
